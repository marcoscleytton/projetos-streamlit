```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("pizzas.csv")

df
```

```python
df.plot(kind="scatter", x="diametro", y="preco",)
```

```python
from sklearn import linear_model

model = linear_model.LinearRegression()
x = df[["diametro"]]
y = df[["preco"]]

model.fit(x, y)
```

```python
print(model.predict([[23]])[0][0])
```
